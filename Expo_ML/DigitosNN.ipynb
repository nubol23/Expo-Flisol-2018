```python
from keras import Sequential
from keras.layers import Dense, Dropout
from sklearn import datasets
from sklearn import model_selection
from keras.utils import to_categorical
import numpy as np
```

```python
data = datasets.load_digits()

X = data.images

Y = data.target

print(X.shape)
print(Y.shape)

m = X.shape[0]
```

```python
X_flattened = X.reshape((m, -1))

Y_encoded = to_categorical(Y)

print(Y[0:15])
print(Y_encoded[0:15,:])
```

```python
X_train, X_test, Y_train, Y_test = model_selection.train_test_split(X_flattened, Y_encoded, test_size=0.20, random_state=1)
print(X_train.shape)
print(Y_train.shape)
print(X_test.shape)
print(Y_test.shape)
```

## Perceptr√≥n multicapa

```python
keras_model = Sequential()
```

```python
keras_model.add(Dense(40, input_dim=64, activation='relu'))
keras_model.add(Dense(25, activation='relu'))
keras_model.add(Dense(10, activation='softmax'))
```

```python
keras_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
keras_model.fit(X_train, Y_train, epochs=2500, batch_size=90)
```

## Cargamos el modelo

```python
from keras.models import model_from_yaml

yaml_file = open('ModeloDigitos.yaml', 'r')
mdl_yaml = yaml_file.read()
yaml_file.close()

mdl = model_from_yaml(mdl_yaml)
mdl.load_weights('ModeloDigitos.h5')

mdl.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
scores = mdl.evaluate(X_test, Y_test)
print('\n' + keras_model.metrics_names[1], scores[1]*100, '%')
```

```python
from sklearn import metrics

Y_decoded = np.argmax(Y_test, axis=1)
pred_decoded = np.argmax(mdl.predict(X_test), axis=1)

print(metrics.classification_report(Y_decoded, pred_decoded))
print('Confussion Matrix')
print(metrics.confusion_matrix(Y_decoded, pred_decoded))
```
